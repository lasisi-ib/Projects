```python
from sklearn.metrics import silhouette_score


model =KMeans(init = 'random',
    n_clusters=4, 
    n_init=500,  # Higher to avoid bad local optima
    max_iter=100,  # Allow more iterations for better convergence
    random_state=100 )

potential_class = model.fit_predict(dt)
sil_score = silhouette_score(dt, model.labels_)
print("Silhouette Score for K=4:", sil_score)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(dt)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=model.labels_, cmap='viridis', alpha=0.6)
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=200, marker='X', c='red')
plt.title("K-Means Clustering Visualization (PCA Reduced)")
plt.show()
```
